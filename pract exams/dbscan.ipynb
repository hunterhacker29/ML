```python
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

# Static input data
X = [[1,1], [2,1], [2,2], [8,8], [8,9], [25,80]]

# Train DBSCAN
# eps = neighborhood distance, min_samples = minimum points to form a cluster
db = DBSCAN(eps=2, min_samples=2)
labels = db.fit_predict(X)

# Plot clusters
colors = ['red','blue','green','purple','orange']
for i in set(labels):
    pts = [X[j] for j in range(len(X)) if labels[j]==i]
    plt.scatter([p[0] for p in pts], [p[1] for p in pts],
                color='black' if i==-1 else colors[i % len(colors)],
                s=100, label=f'Cluster {i}' if i!=-1 else 'Noise')

plt.xlabel('X1'); plt.ylabel('X2'); plt.title('DBSCAN Clustering')
plt.legend(); plt.grid(True); plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Static input data
X = [[1,1], [2,1], [2,2], [8,8], [8,9], [25,80]]

# DBSCAN parameters
db = DBSCAN(eps=2, min_samples=2)
lab = db.fit_predict(X)

# Display clusters and points
cl={}

for i,j in zip(X,lab):
    cl.setdefault(j,[]).append(i)

print(cl)


print(cl)
# Print results
for label, points in cl.items():
    if label == -1:
        print(f"Noise points: {points}")
    else:
        print(f"Cluster {label}: {points}")
```
