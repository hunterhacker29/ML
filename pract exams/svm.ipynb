```python
import matplotlib.pyplot as plt
from sklearn.svm import SVC

# Static data (X1, X2) and labels
X = [[4,1],[4,-1],[6,0],[1,0],[0,1],[0,-1]]
y = [1,1,1,-1,-1,-1]

# Train linear SVM
model = SVC(kernel='linear', C=1000)
model.fit(X, y)

# Plot points
for xi, yi in zip(X, y):
    plt.scatter(xi[0], xi[1], color='blue' if yi==1 else 'red', s=100)

# Plot support vectors
plt.scatter(model.support_vectors_[:,0], model.support_vectors_[:,1],
            s=200, facecolors='none', edgecolors='k')

# Decision boundary
w, b = model.coef_[0], model.intercept_[0]
x = [i/10 for i in range(-10, 71)]
y_boundary = [-(w[0]*xi + b)/w[1] for xi in x]
plt.plot(x, y_boundary, 'k--')

plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Data
x = [[4,1],[4,-1],[6,0],[1,0],[0,1],[0,-1]]
y = [1,1,1,-1,-1,-1]

# Train model
model = SVC()
model.fit(x, y)

# Predict (same data)
y_pred = model.predict(x)

# Accuracy
print("Accuracy:", accuracy_score(y, y_pred))
```
